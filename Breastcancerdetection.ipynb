```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
plt.figure(figsize=(36,14))
sns.heatmap(df.corr(),annot=True , cmap='viridis')
```

```python
x=df.drop(['diagnosis'],axis=1)
y=df['diagnosis']
```

```python
from sklearn.preprocessing import  StandardScaler

scl =StandardScaler()
x_scaled = scl.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
```

```python
log = LogisticRegression(max_iter=1000, solver='lbfgs')
log.fit(x_train, y_train)
```

```python
our_pred= log.predict(x_test)
our_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, our_pred)
```

```python
cm
```

```python
print("num of True Positive :",cm[1][1])
print("num of True Negative :",cm[0][0])
print("num of False Positive :",cm[0][1])
print("num of False Negative :",cm[1][0])
```

```python
from sklearn.metrics import classification_report

cr= classification_report(y_test, our_pred)
print('\nClassification Report for LogisticRegression:')
print(cr)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, our_pred)
print(f'Accuracy of LogisticRegression : {accuracy*100:.2f} %')
```
